Clouds form in clusters. A pixel surrounded by cloud pixels is more likely to be a cloud pixel itself. Likewise, a pixel surround by clear pixels is more likely to be a clear pixel itself. This means that the data is not i.i.d. In general, pixels in an image are not iid. Creating a training set by uniformily randomly picking pixels will result in a traning set that is just a lower res version of the original image. The validation and testing set will also be a low res version of the SAME IMAGE. So in a sense, you're just testing/validating on the same image from the training set.

```python
def split_image(img, n = 2):
    sections = []
    x_del = (img.x.max() + 1 - img.x.min()) / n
    y_del = (img.y.max() + 1 - img.y.min()) / n
    for x in range(n):
        for y in range(n):
            x1 = img.x.min() + x * x_del
            x2 = x1 + x_del
            y1 = img.y.min() + y * y_del
            y2 = y1 + y_del
            sections.append(img[(img.x >= x1) & (img.x < x2) & (img.y >= y1) & (img.y < y2)].copy())
    
    return sections
```

```python
def remove_nonlabeled(df):
    return df[df.label != 0]
```

```python
def merge_frames(frames = []):
    merged = pd.DataFrame()
    for frame in frames:
        merged = merged.append(frame)
    return merged
```

```python
def shuffle_frame(df):
    return df.iloc[np.random.permutation(len(df))]

def split_data_1(images = []):
    sections = [remove_nonlabeled(s) for img in images for s in split_image(img)]
    np.random.shuffle(sections)
    train = shuffle_frame(merge_frames(sections[:7]))
    val = shuffle_frame(merge_frames(sections[7:9]))
    test = shuffle_frame(merge_frames(sections[9:]))
    return [train, val, test] 
```

```python
def split_data_2(images = []):
    merged_image = remove_nonlabeled(merge_frames(images))
    
    # shuffle
    merged_image = shuffle_frame(merged_image)
    
    vi = int(len(merged_image) * 0.58)
    ti = int(len(merged_image) * 0.75)
    return [merged_image.iloc[:vi], 
            merged_image.iloc[vi:ti], 
            merged_image.iloc[ti:]]
```

```python
def baseline_model(X):
    return X.apply(lambda x: -1, axis=1)
```

```python
def validate(yhat, y):
    return (yhat == y).mean()
```

```python
def CVgeneric(model, folds, X, y, loss):
    res = []
    # assign a fold to each row in X
    fold_index = np.repeat(list(range(folds)), math.ceil(len(X) / folds))
    # drop extra rows
    fold_index = fold_index[:len(X)]
    # shuffle the folds
    np.random.shuffle(fold_index)
    
    for i in range(folds):
        X_train = X[fold_index != i]
        y_train = y[fold_index != i]
        X_test = X[fold_index == i]
        y_test = y[fold_index == i]
        
        yhat = model.fit(X_train, y_train).predict(X_test)
        res.append(loss(yhat, y_test))
    
    return res
```

```python
def test_model(model, train_X, train_y, test_X, test_y, folds = 10):
    cv_error = CVgeneric(model, folds, train_X, train_y, validate)
    clf = model.fit(train_X, train_y)
    yhat = clf.predict(test_X)
    return cv_error, validate(yhat, test_y)
```

```python
def plot_curves(curves, labels):
    for i in range(len(curves)):
        fpr, tpr, threshold = curves[i]
        label = labels[i]
        roc_auc = metrics.auc(fpr, tpr)
        cutoff = np.argmax(tpr - fpr)
        plt.plot(fpr, tpr, label = f'{label}: AUC = %0.2f, Cutoff = %0.2f' % (roc_auc, threshold[cutoff]))
        plt.scatter(fpr[cutoff], tpr[cutoff])
    
    plt.legend(loc = 'lower right')
    plt.title('Receiver Operating Characteristic')
    plt.plot([0, 1], [0, 1],'r--')
    plt.xlim([0, 1])
    plt.ylim([0, 1])
    plt.ylabel('True Positive Rate')
    plt.xlabel('False Positive Rate')
```

```python
def calc_roc(model, X, y):
    probs = model.predict_proba(X)
    preds = probs[:,1]
    return metrics.roc_curve(y, preds)
```

```python
def get_XY(df):
    return [df[['NDAI', 'SD', 'CORR', 'DF', 'CF', 'BF', 'AF', 'AN']], df.label]
```

```python
def eval_method(models, train, test):
    train_X, train_y = get_XY(train)
    test_X, test_y = get_XY(test)
    
    cv_errs = []
    test_errs = []
    rocs = []
    
    for model in models:
        # CV error
        cv_err = CVgeneric(model, 10, train_X, train_y, validate)
        
        # Test error
        model = model.fit(train_X, train_y)
        yhat = model.predict(test_X)
        test_err = validate(yhat, test_y)
        
        # ROC curve
        preds = model.predict_proba(test_X)[:, 1]
        roc_auc = metrics.roc_curve(test_y, preds)
        
        cv_errs.append(cv_err)
        test_errs.append(test_err)
        rocs.append(roc_auc)
    
    return np.array(cv_errs), test_errs, rocs
```

```python
from sklearn.tree import DecisionTreeClassifier

n_estimators = 100

X_train, y_train = get_XY(train1.append(val1))
X_test, y_test = get_XY(test1)

dt_stump = DecisionTreeClassifier(max_depth=1, min_samples_leaf=1)
dt_stump.fit(X_train, y_train)
dt_stump_err = 1.0 - dt_stump.score(X_test, y_test)

dt = DecisionTreeClassifier(max_depth=9, min_samples_leaf=1)
dt.fit(X_train, y_train)
dt_err = 1.0 - dt.score(X_test, y_test)

ada_real = AdaBoostClassifier(n_estimators=n_estimators)
ada_real.fit(X_train, y_train)
```

```python
ada_real_err = np.zeros((n_estimators,))
for i, y_pred in enumerate(ada_real.staged_predict(X_test)):
    ada_real_err[i] = 1 - validate(y_pred, y_test)

ada_real_err_train = np.zeros((n_estimators,))
for i, y_pred in enumerate(ada_real.staged_predict(X_train)):
    ada_real_err_train[i] = 1 - validate(y_pred, y_train)
```

```python
plt.figure(figsize=(10, 7))
plt.plot([1, n_estimators], [dt_stump_err] * 2, '-', c='black', label='Decision Stump Test Error')
plt.plot([1, n_estimators], [dt_err] * 2, '--', c='black', label='Decision Tree Test Error')
plt.plot(np.arange(n_estimators) + 1, ada_real_err, '-', c='red', label='AdaBoost Testing Error')
plt.plot(np.arange(n_estimators) + 1, ada_real_err_train, '-', c='blue', label='AdaBoost Training Error')
plt.xlabel('n_estimators')
plt.ylabel('error rate')
plt.title('Error Rate vs Number of Estimators')
plt.legend(loc = 'best')
```

Here is a chart comparing the testing errors of a Decision Stump (max depth = 1), Decision Tree (max depth = 9), and AdaBoost classifier. It's really surprising that the decision stump performed worse than the decision tree. This suggests that this problem is one where it's easy to overfit so it might not be good to use a complex model. This theory is further validated by the testing error of AdaBoost as the number of estimators increases. We can see that the testing error stays pretty constant as the number of estimators increase, and it never performs better than the decision stump. Since AdaBoost increases performance by training new classfiers on previous mistakes, the fact that it doesn't perform better suggests that when the model is wrong, it is wrong in the same way. This suggests that to improve performance, we need better/more data not a better model.

```python
def section_area(model, train):
    X_train, y_train = get_XY(train)
    model.fit(X_train, y_train)

    all_images = [img1, img2, img3]
    sections = [s for img in all_images for s in split_image(img)]

    section_err = []
    err_pix = []
    for section in sections:
        _X = remove_nonlabeled(section)
        X, y = get_XY(remove_nonlabeled(section))
        yhat = model.predict(X)
        section_err.append(validate(y, yhat))
        err_pix.append(_X[yhat != y])

    section_rank = np.argsort(section_err)

    plt.figure(figsize=(10, 10))

    plt.subplot(2, 2, 1)
    plot_img(sections[section_rank[0]])
    er = err_pix[section_rank[0]]
    plt.scatter(er.x, er.y, c='red', s=2)
    plt.title('Accuracy: %0.3f' % section_err[section_rank[0]])

    plt.subplot(2, 2, 2)
    plot_img(sections[section_rank[1]])
    er = err_pix[section_rank[1]]
    plt.scatter(er.x, er.y, c='red', s=2)
    plt.title('Accuracy: %0.3f' % section_err[section_rank[1]])

    plt.subplot(2, 2, 3)
    plot_img(sections[section_rank[-1]])
    er = err_pix[section_rank[-1]]
    plt.scatter(er.x, er.y, c='red', s=2)
    plt.title('Accuracy: %0.3f' % section_err[section_rank[-1]])

    plt.subplot(2, 2, 4)
    plot_img(sections[section_rank[-2]])
    er = err_pix[section_rank[-2]]
    plt.scatter(er.x, er.y, c='red', s=2)
    plt.title('Accuracy: %0.3f' % section_err[section_rank[-2]])

def value_distribution(model, train, test):
    X_train, y_train = get_XY(train)
    model.fit(X_train, y_train)
    X_test, y_test = get_XY(train)
    X_eval = X_test.copy()
    X_eval['correct'] = model.predict(X_test) == y_test
    cols = ['AN', 'DF', 'NDAI']
    
    plt.figure(figsize=(16, 4))
    for i in range(len(cols)):
        plt.subplot(1, 3, i+1)
        col = cols[i]
        sns.distplot(X_eval[X_eval.correct][col], hist=False, label='Correct')
        sns.distplot(X_eval[~X_eval.correct][col], hist=False, label='Incorrect')   
        plt.title(f'Distribution of Values of {col}')
```

Our best classifier is the QDA classfier which achieved an accuracy of 89%. Below are some examples of regions that the classfier performed well on and regions that it performed poorly on. All the pixels that are mixclassified are in red. It appears that the regions the classifier performed well on are uniform--either mostly clear or mostly cloudy. Whereas the regions that the classifier performed poorly on are regions that have more complex cloud coverage. There is a 25%-50% drop in accuracy when classifying complex regions compared to uniform regions. One theory is that uniform regions have values that are more "normal" and have less noise sense the weather pattern is more stable. Complex regions, on the other hand, would have more extreme values / weird value combinations since the weather pattern is more unpredictable. 

We can explore this theory further by looking at the distribution of our 3 features split by when the classifier is correct and when it's incorrect. The correct predictions tend to have higher values of AN and DF and lower values of NDAI. This makes sense as AN and DF represent the brightness of each pixel, so darker clouds can get missclassified. A cloud pixel could be darker than normal due to lower levels of sunlight or a weird angle of reflection due to its shape. We think that this issue can be solved by taking into account the pixels around it, since it is less likely that all the surrounding pixels suffer from the same problem. Furthermore, the classifier tend to be wrong when the values of NDAI are high. According to the paper, larger NDAI values suggest presense of clouds. Since the distribution of NDAI for correct predictions are lower, NDAI seems to be more useful when predicting clear pixels and adds error when predicting clouds. This is especially present in top right region in fig (). An entire block of clear pixels were classified as cloudy. This might be a really high mountain peak which was falsely identified as a cloud due to its high NDAI value. In general, we think that NDAI suffers from the same problems as AN and DF and classification can be improved if we also used the surrounding NDAI values during prediction.

The data doesn't change when we compare the two methods of splitting. This suggests that the error is systematic. We expected method two, which splits the data randomly to perform better. Since the pixels are not i.i.d., the test set from method 2 is less likely to contain new examples. However, since the performance is the same this means that even when trained on very similiar data, the model could not improve its testing accurcy. This suggests that we need more features to improve the accuracy of the model, rather than just a more complex model.
